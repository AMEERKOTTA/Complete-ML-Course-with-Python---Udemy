### REGULARIZATION METHODS

There are three types of Regularization Methods.

+ Ridge Regression  - `l2-regularization`
+ Lasso Regression  - `l1-regularization`
+ Elastic Net Method

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("whitegrid")
```

##### SIMPLE LINEAR REGRESION

```python
from sklearn.linear_model import LinearRegression

## preparing the data
np.random.seed(42)
n_samples = 100
rng = np.random.randn(n_samples) * 10
#print(rng)
## put this rng array as X
X = rng.reshape(-1,1)
#print(X.ndim)
## preparing the y set
y = 0.5 * rng + 2 * np.random.randn(n_samples)
#print(y)

## define the model
model = LinearRegression()
model.fit(X, y)
model_predict = model.predict(X)

## plot the model lines
plt.figure(figsize = (15,8))
plt.scatter(rng, y)
plt.plot(rng, model_predict)
print("The Coefficient Estimates :--", model.coef_)
```

ADDING SOME OUTLIERS

```python
idx = rng.argmax()
y[idx] = 200
idx = rng.argmin()
y[idx] = -200

plt.figure(figsize = (15,8))
plt.scatter(rng, y)
```

```python
## define the model
o_model = LinearRegression()
o_model.fit(X, y)
o_model_predict = o_model.predict(X)

## plot the model lines
plt.figure(figsize = (10,8))
plt.scatter(rng, y)
plt.plot(rng, o_model_predict)
print("The Coefficient Estimates :--", o_model.coef_)
```

### RIDGE REGRESSION

```python
from sklearn.linear_model import Ridge

## define the model
ridge_model = Ridge(alpha = 0.5, normalize = True)
ridge_model.fit(X,y)
ridge_model_predict = ridge_model.predict(X)

## plot the model lines
plt.figure(figsize = (10,8))
plt.scatter(rng, y)
plt.plot(rng, ridge_model_predict)
print("The Coefficient Estimates :--", ridge_model.coef_)
```

### LASSO REGRESSION

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha = 0.5, normalize = True)
lasso_model.fit(X,y)
lasso_model_predict = lasso_model.predict(X)

## plot the model lines
plt.figure(figsize = (10,8))
plt.scatter(rng, y)
plt.plot(rng, lasso_model_predict)
print("The Coefficient Estimates :--", lasso_model.coef_)
```

### ELASTIC NET METHOD

```python
from sklearn.linear_model import ElasticNet

elastic_model = ElasticNet(alpha = 0.5, normalize = True)
elastic_model.fit(X,y)
elastic_model_predict = elastic_model.predict(X)

## plot the model lines
plt.figure(figsize = (10,8))
plt.scatter(rng, y)
plt.plot(rng, elastic_model_predict)
print("The Coefficient Estimates :--", elastic_model.coef_)
```
