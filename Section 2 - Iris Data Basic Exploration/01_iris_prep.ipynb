### IRIS DATA FROM SEABORN

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
iris.describe()
```

```python
iris.info()
```

```python
iris.groupby("species").size()
```

### VISUALIZATION

PAIR PLOT

```python
sns.pairplot(iris, hue = "species", height = 3, aspect = 1)
```

```python
sns.pairplot(iris, hue = "species", diag_kind="kde")
```

HISTOGRAM

```python
iris.hist(bins = 50, figsize = (10,8))
plt.show()
```

```python
iris.hist(edgecolor = "black", linewidth = 1.2, figsize = (10,8))
plt.show()
```

BOXPLOT

```python
iris.boxplot(by = "species", figsize = (10,8))
```

SCATTER PLOT

```python
pd.plotting.scatter_matrix(iris, figsize = (10,8))
plt.show()
```
